```python
import sqlalchemy as db
from sqlalchemy import MetaData, Table, Column, Integer, String, Boolean, Float, ForeignKey
from sqlalchemy.orm import declarative_base
from sqlalchemy.orm import relationship
import pandas as pd
```

```python
Username = 'root'
password = ''
host = ''
database_name = 'gsmarena'
engine = db.create_engine(f'mysql+pymysql://{Username}:{password}@{host}/{database_name}', echo=True)

conn = engine.connect()
Base = declarative_base()
metadata = MetaData()
```

```python
phone_brand = db.Table('phone_brand',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('BrandID', db.Integer, db.ForeignKey('Brand.BrandID')),
)

phone_announce = db.Table( 'phone_announce',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('AnnouncedID', db.Integer, db.ForeignKey('Announced.AnnouncedID')),
)

phone_sim = db.Table('phone_sim',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('SIMID', db.Integer, db.ForeignKey('SIM.SIMID')),
)


phone_sound= db.Table('phone_sound',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('SoundID', db.Integer, db.ForeignKey('Sound.SoundID')),
)


phone_battery = db.Table('phone_battery',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('BatteryID', db.Integer, db.ForeignKey('Battery.BatteryID')),
)

phone_simband = db.Table('phone_simband',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('BandID', db.Integer, db.ForeignKey('SimBand.BandID')),
)

phone_os = db.Table('phone_os',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('OSID', db.Integer, db.ForeignKey('OS.OSID')),
)

phone_cpu = db.Table('phone_cpu',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('CPUID', db.Integer, db.ForeignKey('CPU.CPUID')),
)

phone_gpu = db.Table('phone_gpu',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('GPUID', db.Integer, db.ForeignKey('GPU.GPUID')),
)
phone_ram = db.Table('phone_ram',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('RAMID', db.Integer, db.ForeignKey('RAM.RAMID')),
)
phone_storage= db.Table('phone_storage',Base.metadata ,
    db.Column('PhoneID', db.Integer, db.ForeignKey('Phone.PhoneID')),
    db.Column('STORAGEID', db.Integer, db.ForeignKey('STORAGE.STORAGEID')),
)

class Phone(Base):
    __tablename__ = 'Phone'

    PhoneID = db.Column(db.Integer, primary_key=True, autoincrement=True)
    Name = db.Column(db.String(128))
    Status = db.Column(db.String(128))
    Weight = db.Column(db.Float)
    Length = db.Column(db.Float)
    Width = db.Column(db.Float)
    Diameter = db.Column(db.Float)
    Technology = db.Column(db.String(512))
    Chipset = db.Column(db.String(255))
    Sensors = db.Column(db.String(512))
    DisplaySize = db.Column(db.Float)
    DisplayType = db.Column(db.String(128))
    PPI = db.Column(db.Float)
    ScreenToBodyRatio = db.Column(db.Float)
    ResolutionRatio = db.Column(db.String(128))
    ResolutionPixel = db.Column(db.Float)
    CardSlot = db.Column(db.String(128))
    Bluetooth = db.Column(db.String(128))
    WLAN = db.Column(db.String(128))
    Price = db.Column(db.Float)
    Colors = db.Column(db.String(255))


    brands =relationship('Brand', secondary=phone_brand, back_populates='Phone')
    announces =relationship('Announced', secondary=phone_announce, back_populates='Phone')
    sims =relationship('SIM', secondary=phone_sim, back_populates='Phone')
    sounds =relationship('Sound', secondary=phone_sound, back_populates='Phone')
    batteries =relationship('Battery', secondary=phone_battery, back_populates='Phone')
    bands =relationship('SimBand', secondary=phone_simband, back_populates='Phone')
    OSs =relationship('OS', secondary=phone_os, back_populates='Phone')
    CPUs = relationship('CPU', secondary=phone_cpu, back_populates='Phone')
    GPUs = relationship('GPU', secondary=phone_gpu, back_populates='Phone')
    RAMs = relationship('RAM', secondary=phone_ram, back_populates='Phone')
    storages = relationship('STORAGE', secondary=phone_storage, back_populates='Phone')

class Brand(Base):
    __tablename__ = 'Brand'

    BrandID = db.Column(db.Integer, primary_key=True)
    BrandName = db.Column(db.String(128))
    
    
    phones =relationship('Phone', secondary=phone_brand, back_populates='Brand')



class Announced(Base):
    __tablename__ = 'Announced'

    AnnouncedID = db.Column(db.Integer, primary_key=True)
    Announced_year = db.Column(db.Integer)

    phones =relationship('Phone', secondary=phone_announce, back_populates='Announced')


class SIM(Base):
    __tablename__ = 'SIM'

    SIMID = db.Column(db.Integer, primary_key=True)
    SIMType = db.Column(db.String(128))

    phones =relationship('Phone', secondary=phone_sim, back_populates='SIM')


class Sound(Base):
    __tablename__ = 'Sound'

    SoundID = db.Column(db.Integer, primary_key=True)
    Jack35mm = db.Column(db.Boolean)
    LoudSpeaker = db.Column(db.Boolean)

    phones =relationship('Phone', secondary=phone_sound, back_populates='Sound')


class Battery(Base):
    __tablename__ = 'Battery'

    BatteryID = db.Column(db.Integer, primary_key=True)
    BatteryCapacity = db.Column(db.Integer)

    phones =relationship('Phone', secondary=phone_battery, back_populates='Battery')


class SimBand(Base):
    __tablename__ = 'SimBand'

    BandID = db.Column(db.Integer, primary_key=True)
    _2G = db.Column(db.Boolean)
    _3G = db.Column(db.Boolean)
    _4G = db.Column(db.Boolean)
    _5G = db.Column(db.Boolean)

    phones =relationship('Phone', secondary=phone_simband, back_populates='SimBand')


class OS(Base):
    __tablename__ = 'OS'

    OSID = db.Column(db.Integer, primary_key=True)
    OS = db.Column(db.String(128))

    phones =relationship('Phone', secondary=phone_os, back_populates='OS')


class CPU(Base):
    __tablename__ = 'CPU'

    CPUID = db.Column(db.Integer, primary_key=True)
    CPU = db.Column(db.String(128))

    phones =relationship('Phone', secondary=phone_cpu, back_populates='CPU')


class GPU(Base):
    __tablename__ = 'GPU'

    GPUID = db.Column(db.Integer, primary_key=True)
    GPU = db.Column(db.String(128))

    phones =relationship('Phone', secondary=phone_gpu, back_populates='GPU')
    
class RAM(Base):
    __tablename__ = 'RAM'

    RAMID = db.Column(db.Integer, primary_key=True)
    RAM = db.Column(db.Float)

    phones =relationship('Phone', secondary=phone_ram, back_populates='RAM')


class STORAGE(Base):
    __tablename__ = 'STORAGE'

    STORAGEID = db.Column(db.Integer, primary_key=True)
    STORAGE = db.Column(db.Float)

    phones =relationship('Phone', secondary=phone_storage, back_populates='STORAGE')
```

```python
df= pd.read_csv('.\data_with_out_null.csv')
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
df.dropna(subset=['CPU' , 'Chipset' ,'GPU' , 'Sensors' , 'WLAN'], inplace=True)
df.rename(columns={'Network': 'Technology'}, inplace=True)
df.info()
```

```python
df['Announced'] = df['Announced'].astype('int64')
df['Length'] = pd.to_numeric(df['Length'], errors='coerce')
df = df.dropna(subset=['Length'])
df['Width'] =  pd.to_numeric(df['Width'], errors='coerce')
df['Diameter'] =  pd.to_numeric(df['Diameter'], errors='coerce')

df.dropna(subset=['Diameter'], inplace=True)

df.info()
```

```python
df_phone = df[['name' , 'Status' , 'Weight' , 'Length' , 'Width' , 'Diameter' , 'Technology' ,
                'Chipset' , 'Sensors' , 'Display Size' , 'Display Type' , 'ppi' , 'body ratio' ,
                  'ratio' , 'pixel' , 'Card slot' , 'WLAN' , 'Bluetooth' , 'Price' , 'Colors' ]].copy()

df_phone['PhoneID'] = range(1 , len(df_phone)+1)

df_phone = df_phone[['PhoneID','name' , 'Status' , 'Weight' , 'Length' , 'Width' , 'Diameter' , 'Technology' ,
                      'Chipset' , 'Sensors' , 'Display Size' , 'Display Type' , 'ppi' , 'body ratio' , 'ratio' ,
                        'pixel' , 'Card slot' , 'Bluetooth', 'WLAN'  , 'Price' , 'Colors' ]]

df_phone.rename(columns={'body ratio': 'ScreenToBodyRatio'}, inplace=True)
df_phone.rename(columns={'ratio': 'ResolutionRatio'}, inplace=True)
df_phone.rename(columns={'pixel': 'ResolutionPixel'}, inplace=True)
df_phone.rename(columns={'Display Size': 'DisplaySize'}, inplace=True)
df_phone.rename(columns={'Display Type': 'DisplayType'}, inplace=True)
df_phone.rename(columns={'Card slot': 'CardSlot'}, inplace=True)
```

```python
df_band= df[['2G' , '3G' , '4G' , '5G']].copy()


def create_brand_id(row):
    binary_str = ''.join(map(str, row.astype(int))) 
    brand_id = int(binary_str, 2) + 1 
    return brand_id

df_band['BandID'] = df_band.apply(create_brand_id, axis=1)


df_band = df_band.astype(int)

df_band.drop_duplicates( inplace=True)

df_band.rename(columns={'2G': '_2G'}, inplace=True)
df_band.rename(columns={'3G': '_3G'}, inplace=True)
df_band.rename(columns={'4G': '_4G'}, inplace=True)
df_band.rename(columns={'5G': '_5G'}, inplace=True)


df_band = df_band[['BandID','_2G' , '_3G' , '_4G' , '_5G']]

df_band
```

```python
df_band.to_sql(name = 'simband', con = engine, if_exists='append', index = False)
```

```python
df_sound= df[['Loudspeaker' , '3.5mm jack']].copy()

def create_sound_id(row):
    if row['Loudspeaker'] and row['3.5mm jack']:
        return 1
    elif row['Loudspeaker'] and not row['3.5mm jack']:
        return 2
    elif not row['Loudspeaker'] and row['3.5mm jack']:
        return 3
    else:
        return 4
    

df_sound['soundID'] = df_sound.apply(create_sound_id, axis=1)
df_sound = df_sound.astype(int)


df_sound.rename(columns={'3.5mm jack': 'Jack35mm'}, inplace=True)
df_sound.drop_duplicates( inplace=True)
df_sound = df_sound[['soundID','Loudspeaker' , 'Jack35mm']]

df_sound
```
